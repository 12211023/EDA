```python
from sklearn.impute import KNNImputer

# Select columns for KNN imputation
columns_for_knn = ['Income (USD)', 'Credit Score', 'Property Age', 'Loan Sanction Amount (USD)']


# Check for missing values before imputation
print("Missing values before imputation:")
print(loan_2[columns_for_knn].isnull().sum())
```

```python
# Perform KNN imputation
imputer = KNNImputer(n_neighbors=5)
loan_2_knn_imputed = pd.DataFrame(imputer.fit_transform(loan_2[columns_for_knn]), columns=columns_for_knn)

# Assign imputed values back to the original dataframe
loan_2[columns_for_knn] = loan_2_knn_imputed[columns_for_knn]

# Check for missing values after imputation
print("\nMissing values after imputation:")
print(loan_2[columns_for_knn].isnull().sum())
```

```python
# Calculate and display mean and median before and after imputation
print("\nBefore and After Imputation:")
for col in columns_for_knn:
    mean_before = loan_2[col].mean()
    median_before = loan_2[col].median()
    mean_after = loan_2_knn_imputed[col].mean()
    median_after = loan_2_knn_imputed[col].median()
    print(f"{col}: Mean before = {mean_before}, Median before = {median_before}")
    print(f"{col}: Mean after = {mean_after}, Median after = {median_after}")
```

```python
# Visualize histograms before and after imputation
plt.figure(figsize=(14, 10))
for i, col in enumerate(columns_for_knn, 1):
    plt.subplot(2, 2, i)
    plt.hist(loan_2[col].dropna(), bins=30, color='red', edgecolor='black', alpha=0.7, label='Before Imputation')
    plt.hist(loan_2_knn_imputed[col], bins=30, color='green', edgecolor='black', alpha=0.7, label='After Imputation')
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Check for any remaining missing values in the entire dataframe
print("\nRemaining missing values in the entire dataframe after imputation:")
print(loan_2.isnull().sum())
```

```python
# Function to handle outliers step by step
def handle_outliers(df, column):
    # Step 1: Arrange the data in increasing order
    sorted_data = df[column].sort_values().reset_index(drop=True)
    
    # Step 2: Calculate first (Q1) and third quartile (Q3)
    Q1 = sorted_data.quantile(0.25)
    Q3 = sorted_data.quantile(0.75)
    
    # Step 3: Find interquartile range (IQR)
    IQR = Q3 - Q1
    
    # Step 4: Find lower bound Q1 - 1.5 * IQR
    lower_bound = Q1 - 1.5 * IQR
    
    # Step 5: Find upper bound Q3 + 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Print the calculated values
    print(f'Column: {column}')
    print(f'Q1: {Q1}')
    print(f'Q3: {Q3}')
    print(f'IQR: {IQR}')
    print(f'Lower Bound: {lower_bound}')
    print(f'Upper Bound: {upper_bound}')
    
    # Handle outliers by capping
    df[column] = df[column].apply(lambda x: lower_bound if x < lower_bound else (upper_bound if x > upper_bound else x))

# Handle outliers for each specified column
for col in specific_columns:
    handle_outliers(loan_2, col)

# Check the results
print("\nAfter handling outliers:")
print(loan_2[specific_columns].describe())
```
