```python
import pandas as pd

# Load the dataset

df = pd.read_csv("C:/Users/KRUSHNA KOUSHIK/Desktop/WA_Fn-UseC_-Telco-Customer-Churn.csv")

# Display the first few rows of the dataframe
df.head()
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df['TotalCharges'].fillna(df['TotalCharges'].mean(), inplace=True)
```

```python
df.info()
```

```python
def tenure_group(tenure):
    if tenure <= 12:
        return '0-1 year'
    elif tenure <= 24:
        return '1-2 years'
    elif tenure <= 36:
        return '2-3 years'
    elif tenure <= 48:
        return '3-4 years'
    elif tenure <= 60:
        return '4-5 years'
    else:
        return '5+ years'

df['tenure_group'] = df['tenure'].apply(tenure_group)
df.head()
```

```python
# Transform 'Partner' and 'Dependents' into numerical values
df['Partner'] = df['Partner'].apply(lambda x: 1 if x == 'Yes' else 0)
df['Dependents'] = df['Dependents'].apply(lambda x: 1 if x == 'Yes' else 0)

# Create 'num_family_members' column
df['num_family_members'] = df['Partner'] + df['Dependents']

# Display the first few rows to verify the new column
df.head()
```

```python
df.tail()
```

```python
# Convert "No phone service" to "No" in 'PhoneService' and 'MultipleLines' columns
df['PhoneService'] = df['PhoneService'].replace('No phone service', 'No')
df['MultipleLines'] = df['MultipleLines'].replace('No phone service', 'No')
df.head()
```

```python
# Transform 'PhoneService' and 'MultipleLines' into numerical values
df['PhoneService'] = df['PhoneService'].apply(lambda x: 1 if x == 'Yes' else 0)
df['MultipleLines'] = df['MultipleLines'].apply(lambda x: 1 if x == 'Yes' else 0)
df.head()
```

```python
# Create 'num_phone_services' column
df['num_phone_services'] = df['PhoneService'] + df['MultipleLines']
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df['Churn_numeric'] = df['Churn'].apply(lambda x: 1 if x == 'Yes' else 0)

# Calculate the overall churn rate
churn_rate = df['Churn_numeric'].mean()
print(f"Overall churn rate: {churn_rate:.2%}")
```
