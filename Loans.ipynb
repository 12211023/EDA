```python
import pandas as pd


loan_1 = pd.read_csv('Loan 1.csv')
loan_2 = pd.read_csv('Loan 2.csv')
```

```python
print(loan_1.info())
loan_1.head()
```

```python
loans = pd.concat([loan_1, loan_2], ignore_index=True)
```

```python
loans.info()
```

```python
loans.isnull().sum()
```

```python
import pandas as pd



# Identify non-numeric values in 'Co-Applicant'
non_numeric_co_applicant = loans[~loans['Co-Applicant'].apply(lambda x: str(x).isdigit())]
print("Non-numeric values in 'Co-Applicant':")
print(non_numeric_co_applicant)

# Replace non-numeric values with NaN and then fill or remove them
loans['Co-Applicant'] = pd.to_numeric(loans['Co-Applicant'], errors='coerce')
loans['Co-Applicant'] = loans['Co-Applicant'].fillna(0).astype('int64')  # Assuming '0' for missing values

# Identify non-numeric values in 'Property Price'
non_numeric_property_price = loans[~loans['Property Price'].apply(lambda x: str(x).replace('.', '', 1).isdigit())]
print("Non-numeric values in 'Property Price':")
print(non_numeric_property_price)

# Replace non-numeric values with NaN and then fill or remove them
loans['Property Price'] = pd.to_numeric(loans['Property Price'], errors='coerce')
loans['Property Price'] = loans['Property Price'].fillna(0.0).astype('float64')  # Assuming '0.0' for missing values

# Verify the changes
print("\nUpdated data types for Loan 1:")
print(loans.dtypes)


```

```python
loans.isnull().sum()
```

```python
duplicates = loans[loans.duplicated(subset=['Customer ID'], keep=False)]
print(duplicates)
```

```python
import numpy as np
conditions=[
    (loans['Income Stability'].notnull()),
    (loans['Income Stability'].isnull())&(loans['Profession']=='Commercial associate'),#Low
    (loans['Income Stability'].isnull())&(loans['Profession']=='Working'),#Low
    (loans['Income Stability'].isnull())&(loans['Profession']=='Pensioner'),#High
    (loans['Income Stability'].isnull())&(loans['Profession']=='State servant'),#Low
    (loans['Income Stability'].isnull())&(loans['Profession']=='Unemployed'),#Low
    (loans['Income Stability'].isnull())&(loans['Profession']=='Maternity leave'),#Low
    (loans['Income Stability'].isnull())&(loans['Profession']=='Student'),#Low
    (loans['Income Stability'].isnull())&(loans['Profession']=='Businessman')#Low    
]

values=[
    loans['Income Stability'],
    'Low',
    'Low',
    'High',
    'Low',
    'Low',
    'Low',
    'Low',
    'High'
]
loans['Income Stability'] = np.select(conditions, values, default='Unknown')
loans['Income Stability'].isnull().sum()
```

```python
loans['Gender'].fillna(loans['Gender'].mode()[0], inplace=True)
loans['Gender'].isnull().sum()
```

```python
loans['Has Active Credit Card'].fillna(loans['Has Active Credit Card'].mode()[0], inplace=True)
loans['Has Active Credit Card'].isnull().sum()
```

```python
loans['Property Location'].fillna(loans['Property Location'].mode()[0], inplace=True)
loans['Property Location'].isnull().sum()
```

```python
loans.isnull().sum()
```

```python
# Fill missing values in 'Current Loan Expenses (USD)' with the mean
mean_value = loans['Current Loan Expenses (USD)'].mean()
loans['Current Loan Expenses (USD)'].fillna(mean_value, inplace=True)
loans['Current Loan Expenses (USD)'].isnull().sum()
```

```python
loans['Type of Employment'].fillna(loans['Type of Employment'].mode()[0], inplace=True)
loans['Type of Employment'].isnull().sum()
```

```python
# Fill missing values in 'Current Loan Expenses (USD)' with the mean
median_value = loans['Dependents'].median()
loans['Dependents'].fillna(median_value, inplace=True)
loans['Dependents'].isnull().sum()
```

```python
# Assuming df is your loan dataset loaded properly
import matplotlib.pyplot as plt
# Select columns for KNN imputation
columns_for_knn = ['Income (USD)', 'Credit Score', 'Property Age','Loan Sanction Amount (USD)']

# Visualize histograms before imputation
plt.figure(figsize=(14, 10))
for i, col in enumerate(columns_for_knn, 1):
    plt.subplot(3, 2, i)
    plt.hist(loans[col].dropna(), bins=30, color='blue', edgecolor='black')
    plt.title(f'Histogram of {col} - Before Imputation')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
from sklearn.impute import KNNImputer

# Perform KNN imputation
imputer = KNNImputer(n_neighbors=3)
loans_knn_imputed = pd.DataFrame(imputer.fit_transform(loans[columns_for_knn]), columns=columns_for_knn)

# Visualize histograms after imputation
plt.figure(figsize=(14, 10))
for i, col in enumerate(columns_for_knn, 1):
    plt.subplot(3, 2, i)
    plt.hist(loans_knn_imputed[col], bins=30, color='green', edgecolor='black')
    plt.title(f'Histogram of {col} - After KNN Imputation')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.grid(True)

plt.tight_layout()
plt.show()
```
